```python
#To fetch dataset from the keras library
import tensorflow as tf 
import numpy as np 
import pandas as pd 
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt 
from sklearn.linear_model import SGDClassifier
```

```python
(x_train_digits,y_train_digits),(x_test_digits,y_test_digits)=tf.keras.datasets.mnist.load_data()

(x_train_fashion,y_train_fashion),(x_test_fashion,y_test_fashion)=tf.keras.datasets.fashion_mnist.load_data()

```

```python
import numpy as np

unique, counts = np.unique(y_train_digits, return_counts=True)
plt.bar(unique, counts)
plt.xticks(unique)
plt.xlabel("Label")
plt.ylabel("Quantity")
plt.title("Labels in MNIST Training dataset(digits)")

```

```python
print(x_train_digits.shape,y_train_digits.shape)

plt.imshow(x_train_digits[0])
plt.title(y_train_digits[0])
```

```python
sgd_cls=SGDClassifier(loss="log_loss")
sgd_cls.fit(x_train_digits.reshape(60000,28*28),y_train_digits)

sdg_cls=SGDClassifier()
sdg_cls.fit(x_train_fashion.reshape(60000,28*28),y_train_fashion)
```

```python
#please integrate your code her for MnistFashion
y_pred=sgd_cls.predict(x_test_digits.reshape(10000,784))
#actual labels vs predicted labels 
print("actual: ",y_test_digits,"predicted: ",y_pred)
```

```python
cm=confusion_matrix(y_test_digits,y_pred)
sns.heatmap(cm)
plt.xlabel("predicted")
plt.ylabel("actual")
```

```python
print(classification_report(y_test_digits,y_pred))
```
